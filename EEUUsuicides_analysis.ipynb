```python
# Importing + csv reference
import pandas as pd
import matplotlib.pyplot as plt

suicides = pd.read_csv("Suicides by race EEUU.csv")
```

#### Female Suicide Trend through years (No race filter)

```python
# 1) Get the total suicides of each year of women:
female_suicides_all_ages = suicides[(suicides["STUB_LABEL"] == "Female") & (suicides["AGE"] == "All ages") & (suicides["YEAR"].between(1980, 2018))]
female_suicides = female_suicides_all_ages["ESTIMATE"]

# 2) get the years
years = female_suicides_all_ages["YEAR"]

# 3) plot it in a bar graph
plt.bar(years, female_suicides, color="orange")
plt.ylabel("Suicides")
plt.xlabel("Years")
plt.title("USA Female Suicides per Year (estimate)")
plt.show()
```

### Total White vs Black Male Suicides

```python
# Get total white male suicides: (1980-2020)
white_male_suicides = suicides[(suicides["STUB_LABEL"] == "Male: White") & (suicides["YEAR"].between(1980, 2020))]
totalwhite = white_male_suicides["ESTIMATE"].sum()

# Get total black male suicides (1980-2020)
black_male_suicides = suicides[(suicides["STUB_LABEL"] == "Male: Black or African American") & (suicides["YEAR"].between(1980, 2020))]
totalblack = black_male_suicides["ESTIMATE"].sum()

# Prepare Data:
suicideswb_list = []
suicideswb_list.append(totalwhite)
suicideswb_list.append(totalblack)
roles = ["White", "Black"]

# Plot in a Pie Chart
plt.pie(suicideswb_list, labels=roles, autopct="%1.0f%%", explode=[0.1, 0])
plt.title("White vs Black Male suicide Comparison between 1980-2020")
plt.show()
```
